## Escapes


<br>.       - Any Character Except New Line
<br>\d      - Digit (0-9)
<br>\D      - Not a Digit (0-9)
<br>\w      - Word Character (a-z, A-Z, 0-9, _): common search for a work character, lower case letter, digit and underscore
<br>\W      - Not a Word Character
<br>\s      - Whitespace (space, tab, newline)
<br>\S      - Not Whitespace (space, tab, newline)

### Anchors: use in conjunction with other patterns
<br>\b      - Word Boundary: whitespace and the start of a line
<br>\B      - Not a Word Boundary
<br>^       - Beginning of a String
<br>$       - End of a String (end of a quotation)
<br>[]      - Matches Characters in brackets, only one character within the group is matched 
<br>[^ ]    - Matches Characters NOT in brackets, negation
<br>|       - Either Or
<br>( )     - Group

<br> () represents patterns within a group, [] only selects one character within the brackets

### Quantifiers (match multiple characters at a time, used AFTER the character or character group):
<br> quantifiers only applies to one character or character group
<br>*       - 0 or More
<br>+       - 1 or More
<br>?       - 0 or One
<br>{3}     - Exact Number
<br>{3,4}   - Range of Numbers (Minimum, Maximum)


#### Sample Regexs ####

```python
pattern = re.compile(r'\d{3}.\d{3}.\d{4}')
with open('data.txt') as f:
    contents = f.read()
    matches=pattern.finditer(contents)
    
    for match in matches:
        print(match)
```

```python
#match 800 and 900 numbers
pattern = re.compile(r'[89]00.\d{3}.\d{4}')
with open('data.txt') as f:
    contents = f.read()
    matches=pattern.finditer(contents)
    
    for match in matches:
        print(match)
```

```python
#a-z
pattern = re.compile(r'[o-zO-Z]')
with open('data.txt') as f:
    contents = f.read()
    matches=pattern.finditer(contents)
    
    for match in matches:
        print(match)
```

```python
#^ caret at the beginning represents the beginning of a text
#^ caret at the beginning of the character set within the bracket is a negation to the character set
pattern = re.compile(r'[^/b]at')
with open('data.txt') as f:
    contents = f.read()
    matches=pattern.finditer(contents)
    
    for match in matches:
        print(match)
```

```python
pattern = re.compile(r'Mr\.?\s\w*')
matches=pattern.finditer(text_to_search)

for match in matches:
    print(match)
```

```python
#group: uses a parentesis
#within the parentesis, we can match either pattern within the group
pattern = re.compile(r'M(r|rs|s)\.?\s\w*')
matches=pattern.finditer(text_to_search)

for match in matches:
    print(match)
```

```python
pattern = re.compile(r'[a-zA-Z0-9.-]+@[a-zA-Z-]+\.(com|edu|net)')
with open('emails.txt') as f:
    contents = f.read()
    matches=pattern.finditer(contents)
    
    for match in matches:
        print(match)
```

```python
pattern = re.compile(r'https?://(www\.)?\w+\.\w+')
with open('urls.txt') as f:
    contents = f.read()
    matches=pattern.finditer(contents)
    
    for match in matches:
        print(match)
```

```python
#capture only the domain
#we have three different groups
pattern = re.compile(r'https?://(www\.)?(\w+)(\.\w+)')
with open('urls.txt') as f:
    contents = f.read()
    matches=pattern.finditer(contents)
    #the match has a group 
    for match in matches:
        #group 0 is the entire url
        print(match.group(0))
        #group 1 is the www if exists
        print(match.group(1))
        #group 2 is the domain name
        print(match.group(2))
        #group 3 is the .com or .gov
        print(match.group(3))
```

```python
# sub urls uses the pattern to substitue out group 2 and group 3 for all the matches in urls
#every time it finds a match, it will substitue the match with group 2 and group 3
subbed_urls=pattern.sub(r'\2\3',contents)
print(subbed_urls)
```

```python
pattern=re.compile(r'(Mr|Mrs|Ms)\.?\s\w*')

matches=pattern.finditer(text_to_search)
for match in matches:
    print(match)
```

```python
pattern=re.compile(r'(Mr|Mrs|Ms)\.?\s\w*')

matches=pattern.finditer(text_to_search)
for match in matches:
    print(match.group(0))
```

```python
#findall() returns the matches as a list of string
# if it is matching groups, it will only return the groups

pattern=re.compile(r'(Mr|Mrs|Ms)\.?\s\w*')

matches=pattern.findall(text_to_search)
for match in matches:
    print(match)
    
#it only returns the first group
```

```python
#match() returns the first match, if there isn't a match, it returns a none
matches=pattern.match(text_to_search)
print(matches)
```

```python
#search() search the entire string, prints out the search that it finds
#if it doesn't exist, None is returned
matches=pattern.search(text_to_search)
print(matches)
```
